### Application Programming Interface (API)
- a way for two or more computer programs to communicate with each other
- example: ordering from KFC
  - you (client): make request to the source(kitchen) through API (crew) on what you what provided the rules (menu)
  - crew (API): process the request and deliver the results
  - menu (API rules/ specifications): rules/ specifications for getting the things needed
  - kitchen (source): data source

### API
- API is like a waiter at a restaurant who takes your order and brings it to the kitchen
- "restaurant" is a software application
- API receives requests from one software component (the client) and delivers them to another software component (the server) in a format that both can understand

### Representational State Transfer API (REST API)
- REST API is a type of API that uses a set of rules to make it easier for developers to build and use APIs
- it like a common language that everyone agrees to speak so that they can work together more easily
- developers can use a set of standard commands to request data or perform actions on a server, and the server will respond in a predictable way.
- example
  - to search for movie titles
  - to request a list of movies that match your search criteria, and the server would respond with a list of movie titles and other information

### SUMMARY
Overall, APIs and REST APIs are like communication channels that allow different software components to work together more easily, like waiters in a restaurant or people speaking a common language.

### HyperText Transfer Protocol (HTTP)
- rules/ protocol to follow in order to make request from web server

### HTTP Status Code
- 1xx: Informational
- 2xx: Success
- 3xx: Redirection
- 4xx: Client Error
- 5xx: Server Error

### API Parameters

- input to be specified, like ordering in KFC, we need to provide our order list so that the source know what we want
- the parameters same as the function case where we only required to provide those are required, others are optional where default values have been provided


image.png

```python
# to specify the parameter
parameters = {
    'lat': 3.415374,
    'lng': 102.033276,
    'formatted': 0,
}
```

```python
sun_api = requests.get(url='https://api.sunrise-sunset.org/json', params=parameters)
sun_api_data = sun_api.json()
sunrise_hour = sun_api_data['results']['sunrise'].split("T")[1].split(":")[0]
sunset_hour = sun_api_data['results']['sunset'].split("T")[1].split(":")[0]
sunrise_min = sun_api_data['results']['sunrise'].split("T")[1].split(":")[1]
sunset_min = sun_api_data['results']['sunset'].split("T")[1].split(":")[1]
print(sunrise_hour, sunrise_min)
print(sunset_hour, sunset_min)
```
